# **Graficación de interpolaciones lineales con los datos medidos de cada material a partir de la ecuación de dilatación lineal térmica**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos de entrada
materiales = ["Bronce", "Zinc", "Aluminio"]
delta_L = np.array([1.63, 2.21, 1.90])  # en mm
delta_T = np.array([55.85, 58.50, 53.78])  # en °C
L0 = 750  # Longitud inicial en mm
alpha = np.array([38.91, 54.79, 43.30]) * 1e-6  # Coeficientes de dilatación en °C^-1
```

```python
# Generar gráficos para cada material por separado

for i, material in enumerate(materiales):
    plt.figure(figsize=(10, 10))
    # Crear un rango de temperaturas alrededor de ΔT
    temp_range = np.linspace(delta_T[i] - 10, delta_T[i] + 10, 100)
    # Calcular ΔL usando la fórmula de dilatación térmica
    delta_L_range = alpha[i] * L0 * temp_range
    plt.plot(temp_range, delta_L_range, label=f'{material} Ajuste Lineal', color='blue')
    plt.scatter(delta_T[i], delta_L[i], color='red', label='Punto de Dato')
    plt.title(f'{material}: ΔL vs ΔT')
    plt.xlabel('Diferencia de Temperatura (°C)')
    plt.ylabel('Diferencia de Longitud (mm)')
    plt.grid(True)
    plt.legend()
    plt.show()
```
