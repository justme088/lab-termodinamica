# **Obtención de valores estimados de resistencia en función de temperatura a través de una interpolación polinomial cúbica**

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
# Datos de temperatura y resistencia
temperaturas = list(range(-10, 101))
resistencias = [
    611870, 577940, 546070, 516130, 488000, 461550, 436680, 413280, 391270, 370540, 351020, 332640, 315320, 298990,
    283600, 269080, 255380, 242460, 230260, 218730, 207850, 197560, 187840, 178650, 169950, 161730, 153950, 146580,
    139610, 133000, 126740, 120810, 115190, 109850, 104800, 100000, 95447, 91126, 87022, 83124, 79422, 75903, 72560,
    69380, 66356, 63480, 60743, 58138, 55658, 53297, 51048, 48905, 46863, 44917, 43062, 41305, 39605, 37995, 36458,
    34991, 33591, 32253, 30976, 29756, 28590, 27475, 26409, 25390, 24415, 23483, 22590, 21736, 20919, 20136, 19386,
    18668, 17980, 17321, 16689, 16083, 15502, 14945, 14410, 13897, 13405, 12932, 12479, 12043, 11625, 11223, 10837,
    10467, 10110, 9767.2, 9437.7, 9120.8, 8816.0, 8522.7, 8240.6, 7969.1, 7707.7, 7456.2, 7214.0, 6980.6, 6755.9,
    6539.4, 6330.8, 6129.8, 5936.1, 5750.1, 5569.3
]

# Crear el modelo de interpolación
interpolacion = interp1d(temperaturas, resistencias, kind='cubic')

# Función para obtener la resistencia a una temperatura deseada
def obtener_resistencia(temperatura):
    return float(interpolacion(temperatura))

# Visualización de la interpolación
temp_range = np.linspace(min(temperaturas), max(temperaturas), 500)
res_range = interpolacion(temp_range)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(temperaturas, resistencias, 'o', label='Datos Originales')
plt.plot(temp_range, res_range, '-', label='Interpolación polinomial (cúbica)')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Resistencia (Ω)')
plt.legend()
plt.title('Interpolación de Resistencia en Función de la Temperatura')
plt.grid(True)
plt.show()
```

```python
print("Para el material 1 (bronce):")
print("Resistencia inicial:", round(obtener_resistencia(24.50),2),"Ω")
print("Resistencia final:", round(obtener_resistencia(80.35),2),"Ω")
print()
print("Para el material 2 (aluminio):")
print("Resistencia inicial:", round(obtener_resistencia(25.27),2),"Ω")
print("Resistencia final:", round(obtener_resistencia(83.77),2),"Ω")
print()
print("Para el material 3 (zinc):")
print("Resistencia inicial:", round(obtener_resistencia(24.80),2),"Ω")
print("Resistencia final:", round(obtener_resistencia(78.58),2),"Ω")
```
